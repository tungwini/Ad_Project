```python
Y=df['target']
```

```python
X=df[:]
X.drop(['target','Unnamed: 0'],axis=1,inplace=True)
```

```python
X_us=df.loc[df['birth_country']==' United-States',:]
Y_us=X_us['target']
X_us.drop(['target','Unnamed: 0'],axis=1,inplace=True)
X_oc=df.loc[df['birth_country']!=' United-States',:]
Y_oc=X_oc['target']
X_oc.drop(['target','Unnamed: 0'],axis=1,inplace=True)
```

```python
from sklearn import preprocessing
var_trans_us={}
for cols in X_us.columns:
    try:
        if X_us[cols].dtype == object:
            var_trans_us[cols]=preprocessing.LabelEncoder()
            X_us[cols]=var_trans_us[cols].fit_transform(X_us)
    except:
        print cols
        X_us.drop([cols],axis=1,inplace=True)
var_trans_oc={}
for cols in X_oc.columns:
    try:
        if X_oc[cols].dtype == object:
            var_trans_oc[cols]=preprocessing.LabelEncoder()
            X_oc[cols]=var_trans_oc[cols].fit_transform(X_oc)
    except:
        print cols
        X_oc.drop([cols],axis=1,inplace=True)
```

```python
df.shape
```

```python
from sklearn.cross_validation import train_test_split
X_us_train,X_us_test,Y_us_train,Y_us_test =train_test_split(X_us, Y_us, test_size=0.20, random_state=4234)
X_oc_train,X_oc_test,Y_oc_train,Y_oc_test =train_test_split(X_oc, Y_oc, test_size=0.20, random_state=42235)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
k=8
knn_model = KNeighborsClassifier(n_neighbors=k)
knn_model.fit(X_us_train,np.ravel(Y_us_train))
#acc=acc.append(pd.DataFrame([[k,knn_model.score(X_test,np.ravel(Y_test)),np.nan]],columns=['k','knn_acc','logit_acc']))
print "K="+str(k)+ " " + "Accuracy="+str(knn_model.score(X_us_test,np.ravel(Y_us_test)))

k=8
knn_model = KNeighborsClassifier(n_neighbors=k)
knn_model.fit(X_oc_train,np.ravel(Y_oc_train))
#acc=acc.append(pd.DataFrame([[k,knn_model.score(X_test,np.ravel(Y_test)),np.nan]],columns=['k','knn_acc','logit_acc']))
print "K="+str(k)+ " " + "Accuracy="+str(knn_model.score(X_oc_test,np.ravel(Y_oc_test)))
```
